```python
import tensorflow as tf
print(tf.__version__)

# Model টা ONNX এ convert করি
import tf2onnx
import numpy as np

model = tf.keras.models.load_model("butterfly_model.keras")

input_signature = [tf.TensorSpec([1, 150, 150, 3], tf.float32)]
model_proto, _ = tf2onnx.convert.from_keras(model, input_signature=input_signature)

with open("butterfly_model.onnx", "wb") as f:
    f.write(model_proto.SerializeToString())

print("✅ ONNX model saved!")
```

```python
import tf2onnx
import tensorflow as tf

model = tf.keras.models.load_model("butterfly_model.keras")

# SavedModel হিসেবে আগে save করুন
model.export("saved_model_dir")

# তারপর SavedModel থেকে ONNX convert করুন
input_signature = [tf.TensorSpec([1, 150, 150, 3], tf.float32, name="input")]

model_proto, _ = tf2onnx.convert.from_saved_model(
    "saved_model_dir",
    input_names=["input"],
    output_names=["output"],
    opset=13
)

with open("butterfly_model.onnx", "wb") as f:
    f.write(model_proto.SerializeToString())

print("✅ Done!")
```

```python
import tensorflow as tf

model = tf.keras.models.load_model("butterfly_model.keras")

# TFLite format এ convert করুন (সবচেয়ে simple)
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open("butterfly_model.tflite", "wb") as f:
    f.write(tflite_model)

print(f"✅ TFLite model saved! Size: {len(tflite_model)/1024/1024:.1f} MB")
```
